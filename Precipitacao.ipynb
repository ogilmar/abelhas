```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
dados = {
    'Data': ["Março","Abril","Maio","Junho","Julho","Agosto","Setembro","Outubro","Novembro","Dezembro"],
    'Precipitação': [594.8,481.2,274.4,185.4,110.2,15,7,9,25.8,19]
}
```

```python
df = pd.DataFrame(dados)
df
```

```python
plt.figure(figsize=(10, 5))  # Define o tamanho da figura
plt.plot(df['Data'], df['Precipitação'], marker='o', color='blue', linestyle='-', linewidth=2, markersize=8)
plt.title('Precipitação ao longo do tempo', fontsize=16)
plt.xlabel('Mês', fontsize=12)
plt.ylabel('Precipitação (mm)', fontsize=12)

caminho_pasta = 'C:/Users/Usuario/Documents/Documentos_pessoais/Estudos_abelha_solitaria'  # Substitua pelo caminho da sua pasta
nome_arquivo = 'grafico_precipitacao.png'
caminho_completo = os.path.join(caminho_pasta, nome_arquivo)
plt.savefig(caminho_completo, dpi=300, bbox_inches='tight')  # dpi=300 para alta qualidade
print(f"Gráfico salvo em: {caminho_completo}")
```
